```python
import os
import random
import sqlite3
import os.path

dbname='BPA.db'
tablename='USER'
datatable='SERVICES'

def createDBTable():

    conn = sqlite3.connect(dbname)
    cursor = conn.cursor()
    cursor.execute(f"DROP TABLE IF EXISTS {tablename}")
    query = f"""CREATE TABLE {tablename}(
            id INT PRIMARY KEY NOT NULL,
            name CHAR(20) NOT NULL, 
            password CHAR(20)NOT NULL, 
            roll CHAR(20) NOT NULL, 
            location CHAR(20) NOT NULL)"""
    cursor.execute(query)
    conn.commit()
    print('user table')
    query = f"""CREATE TABLE {datatable}(
            pid INT PRIMARY KEY NOT NULL,
            service_name CHAR(20) NOT NULL, 
            price INT NOT NULL, 
            location CHAR(20) NOT NULL)
            """
    cursor.execute(query)
    conn.commit()
    conn.close()
    print("Beauty Parlor Database is Constructed")


print('-'*48,"\nWelcome to the Beauty Parlor Aggregator")

def SignUp():
    rid = random.randint(1001, 5000)
    print('----Welcome User----')
    name = input("Enter name:")
    password = input("Enter password:")
    roll = input("Enter service provider/customer:")
    location = input("Enter location:")
    conn = sqlite3.connect(dbname)
    query = (f'INSERT INTO {tablename} (id,name,password,roll,location) '
             'VALUES (:id, :name, :password, :roll, :location);')
    params = {
        'id': rid,
        'name': name,
        'password': password,
        'roll': roll,
        'location': location
    }
    try:
        conn.execute(query, params)
        print("Details added")
        print(f"Your Id is:{rid}")
    except Exception as e:
        print("Oops An Error Occured", e)
    finally:
        conn.commit()
        conn.close()

def SignIn():
    print('----Welcome Back,----')
    rid=int(input("Enter Your UserId:"))
    password=input("Enter Your Password")
    conn = sqlite3.connect(dbname)
    query=(f'select * from {tablename} where id={rid}')
    cursor=conn.execute(query)
    datalist=cursor.fetchall()
    if len(datalist)>0:
        for item in datalist:
            if item[3]=='customer':
                CustomerMenu(item[1])
            elif item[3]=='provider':
                ProviderMenu(item[1],item[0])
    else:
        print('SignUp before SignIn, Thank You')

def CustomerMenu(CustomerName):
    print('-'*48)
    print(f'----Customer Menu----\nHello {CustomerName}')
    print('-' * 48,"\nMenu")
    choice = ['Main Menu', 'Search For Service']
    print("".join([str(id+1) + ". " + val + "\n" for id, val in enumerate(choice)]))

    option = eval(input("Select your option please:"))
    print('-' * 48)
    funcdict = {2:Search}
    if option == 1:
        Menu()
    else:
        funcdict[option](CustomerName)
    
def ProviderMenu(ProviderName,ProviderId):
    print('-'*48)
    print(f'----Provider Menu----\nHello {ProviderName}')
    print('-' * 48,"\nMenu")
    choice = ['Main Menu', 'Upload Service Details']
    print("".join([str(id+1) + ". " + val + "\n" for id, val in enumerate(choice)]))

    option = eval(input("Select your option please:"))
    print('-' * 48)
    funcdict = {2:Upload}
    if option == 1:
        Menu()
    else:
        funcdict[option](ProviderName)
        
def Upload(ProviderName):
    pid = random.randint(0, 1000)
    print(f'Upload Your Services Mr/Mrs. {ProviderName}')
    service_name = input("Enter service name:")
    price = int(input("Enter price of the service:"))
    location = input("Enter location of the servvice:")
    conn = sqlite3.connect(dbname)
    query = (f'INSERT INTO {datatable} (pid,service_name,price,location) '
             'VALUES (:pid, :service_name, :price, :location);')
    params = {
        'pid': pid,
        'service_name': service_name,
        'price': price,
        'location': location
    }
    try:
        conn.execute(query, params)
        print("Service added")
        print(f"Your ServiceId is:{pid}")
    except Exception as e:
        print("Oops An Error Occured while uploadding your service", e)
    finally:
        conn.commit()
        conn.close()

def Search(CustomerName):
    print(f'Search Your Services Here, Mr/Mrs.{CustomerName}')

def Menu():
    print('-' * 48,"\nMenu")
    choice = ['Exit', 'SignIn', 'SignUp']
    print("".join([str(id ) + ". " + val + "\n" for id, val in enumerate(choice)]))
    option = eval(input("Select your option please:"))
    print('-' * 48)
    return option

while True:
    if not os.path.isfile(dbname): createDBTable() # Create DB and Table if does not exist
    option = Menu()
    funcdict = {1:SignIn,2:SignUp}
    if option == 0:
        break
    else:
        funcdict[option]()
```
